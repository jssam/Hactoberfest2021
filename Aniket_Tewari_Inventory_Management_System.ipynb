**Importing json Module**

```python
import json
```

**Reading Recod.json File**

```python
fd=open("record.json",'r')
data=fd.read()
fd.close()
records=json.loads(data)
```

```python
records
```

**Purchasing The Product**

```python
prodID=str(input("Enter the Product - Id= "))
quantity=int(input("Enter the Quantity= "))

if(quantity<records[prodID]['Quantity']):
  print("*****************************")
  print("Hey, Welcome!! \nYou Chose--\n")
  print("Product ID= ",prodID)
  print("Product Name= ", records[prodID]['Name'])
  print("Product Amount= ", records[prodID]['Price'])
  print("Manufacturing Date= ", records[prodID]['Manufacturing Date'])
  print("Expiry Date= ", records[prodID]['Expiry Date'])
  print("Available Quantities= ",records[prodID]['Quantity'])
  print("Discount= ",str(records[prodID]['Discount']) + " %")
  print("****************************\n")
  totalDis=int(records[prodID]['Discount'])*quantity
  dis=totalDis*int(records[prodID]['Price'])*quantity*0.01
  print("Amount= ",(records[prodID]['Price'])*quantity)
  print("Total Discount you got= ",str(quantity*int(records[prodID]['Discount'])) + " %")
  print("Discount Amount= ",dis)
  print("****************************")
  print("\nTotal Amount to be paid= ", str(records[prodID]['Price']*quantity-dis)  + " /-")

  records[prodID]['Quantity'] = records[prodID]['Quantity']-quantity
else:
  print("*****************************")
  print("Hey, Welcome!! \nYou Chose--\n")
  print("Product ID= ",prodID)
  print("Product Name= ", records[prodID]['Name'])
  print("Product Amount= ", records[prodID]['Price'])
  print("Manufacturing Date= ", records[prodID]['Manufacturing Date'])
  print("Expiry Date= ", records[prodID]['Expiry Date'])
  print("Available Quantities= ",records[prodID]['Quantity'])
  print("Discount= ",str(records[prodID]['Discount']) + " %")
  print("******************************\n")
  print("OOps, no sufficient quantities of the selected item are available.")
```

**Updating the Recod.json File**

```python
js=json.dumps(records)
fd=open("record.json",'w')
fd.write(js)
fd.close()
```

**Creating Sales.json File for storing the Sales Data.**

```python
fd=open("sales.json",'r')
data=fd.read()
sales=json.loads(data)
sales[prodID]={"Details" : {'Product ID' : prodID, 'Name':records[prodID]['Name'], 'Quantity' : quantity, 'Discount': records[prodID]['Discount'], 'Amount': records[prodID]['Price']*quantity-dis}}

js = json.dumps(sales)
fd = open("sales.json",'w')
fd.write(js)
fd.close()
```
